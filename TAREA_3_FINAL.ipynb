## Estenografía

La palabra esteganografía se deriva de las palabras griegas "stegos" que significan "cubierta" y "grafía" que significa "escritura" y la define como "escritura cubierta"[1]. La esteganografía y la criptografía están estrechamente relacionadas. La criptografía codifica los mensajes para que no puedan ser entendidos. La esteganografía, por otro lado, esconderá el mensaje para que no haya conocimiento de la existencia del mensaje en primer lugar. En algunas situaciones, el envío de un mensaje cifrado despertará sospechas, mientras que un mensaje "invisible" no lo hará. Ambas ciencias pueden combinarse para producir una mejor protección del mensaje. En este caso, cuando la esteganografía falla y el mensaje puede ser detectado, sigue siendo inútil, ya que está encriptado mediante técnicas de criptografía. [2] La Escenografía se compone de dos elementos principales el mensaje y el portador o la cobertura del mensaje. Se puede clasificar mediante el tipo de portador o cobertura que se utilice para portar un mensaje. Las principales clasificaciones son Texto, Audio e Imagenes [3], [4]:

* Texto:
El método de esteganografía utiliza los medios de texto para ocultar los datos es conocidos como esteganografía de texto. Existen diferentes técnicas para incrustar los datos secretos en archivos de texto.
  * Método basado en el formato: Este método modifica el texto existente para ocultar los datos de tal manera que implica la inserción de espacios, redimensionar el texto, cambiar el estilo del texto.

  * Método aleatorio y estadístico: En este método se ocultan los caracteres que aparecen en secuencia aleatoria. El método estadístico determina las estadísticas tales como la media, la varianza y el texto chi cuadrado que miden la cantidad de información redundante que se debe ocultar dentro del texto.

  * Método Lingüístico: Es la combinación de sintaxis y semántica. La estegoanálisis sintáctica asegura la estructura correcta ya que el texto se genera a partir de la gramática. En el método semántico se asigna un valor a los sinónimos y los datos pueden ser codificados a la palabra real del texto. 

* Imagenes: 
En este método, las imágenes se utilizan como objeto de cobertura. La imagen Esteganografía, se puede clasificar en diferentes categorías.

  * Dominio espacial: En el ámbito espacial, cubra la imagen y los datos secretos modificados mediante la codificación LSB y de nivel. Primero, la imagen de la portada se descompone en planos de bits y luego LSB es de planos de bits reemplazados por el ajuste de datos secretos. Este método proporciona una mejor calidad de imagen. El único inconveniente de la inserción de LSB es la simplicidad del proceso de extracción.

  * Dominio de frecuencia: Los datos secretos se ocultan en áreas significativas de la imagen cubierta, lo que hace que los datos sean más vigorosos para ataques como la compresión, el recorte o los métodos de procesamiento de imágenes que el enfoque LSB. Este método de transformación incluye DCT, DWT y DFT.
 
  * Dominio adaptativo: Este método de esteganografía es un caso especial de dos métodos: dominio espacial y dominio de transformación. También se conoce como "Statistics Aware Embedding" y "Masking". Las características globales de las imágenes se utilizan antes de integrar datos secretos en los coeficientes de DCT o DWT. Estas estadísticas decidirán dónde se pueden hacer los cambios. 

* Audio: 
La técnica se conoce como esteganografía de audio. Este método incorpora el mensaje secreto en los archivos de sonido WAV, AU y MP3. Existen diferentes métodos a través de los cuales se explora la esteganografía de audio:

  * Codificación de bits bajos: Este método se utiliza para la predicción del período de afinación que se lleva a cabo durante la codificación de voz de bajo nivel de bits. De esta manera, se mantiene la sincronización entre la ocultación de información y la codificación del habla. 

  * Codificación de fase: En este método, el archivo stream divide el audio en bloques e integra toda la secuencia secreta en el espectro de fases del primer bloque. 

  * Espectro ensanchado: Un método particular de codificación de espectro ensanchado es el DSSS (Direct Sequence Spread Spectrum) que extiende la esteganografía multiplicándose por cierta secuencia pseudoaleatoria. 

Las aplicaciones son muy diversas, esta técnica a sido utilizada por distintos tipos de organizaciones tanto gubernamentales como terroristas, con el objetivo de transmitir mensajes.

```
#Codigo python
from IPython.display import Image
import requests
from stegano import lsb

#Se descarga una imagen.
url_imagen = "https://i.dlpng.com/static/png/5500153-hd-ny-skyline-png-view-of-new-york-free-unlimited-download-new-york-skyline-png-1502_493_preview.png" # El link de la imagen
nombre_local_imagen = "ny.png" # El nombre con el que queremos guardarla
imagen = requests.get(url_imagen).content
with open(nombre_local_imagen, 'wb') as handler:
	handler.write(imagen)
print('Imagen Descargada')
Image('ny.png')

```

```
#Parte 1:
# Se generan un par de Claves RSA (Publica y Privada)
from Crypto.PublicKey import RSA
rsa_key_pair = RSA.generate(4096, e=65537)
private_key = rsa_key_pair.exportKey("PEM")
print('private_key',private_key)
pub_key = rsa_key_pair.publickey()
public_key = pub_key.exportKey("PEM")
print('public_key',public_key)

```

```
#Parte 2:
# Se aplica Steganografia para ocultar la clave publica.
data_to_hide = public_key
original_image="./ny.png"
secret = lsb.hide(original_image, data_to_hide)
secret.save("./stegano-ny.png")
print('Imagen con Clave RSA Publica oculta. ')
Image('stegano-ny.png')
```

```
#Parte 3:
# Se obtiene la Clave RSA publica de la Imagen.
clear_message = lsb.reveal("./stegano-ny.png")

print('Mensaje',clear_message)
```
