## One Time Pad
En el contexto de la Criptografía existen muchos algoritmos y estrategias para realizar cifrado de la información con el objetivo de que la informacion permanezca segura. One time Pad es una de las técnicas más conocidas y famosas para realizar el cifrado de la información ya que posee caracteristicas que son únicas.


En criptografía, One Time Pad se define como una técnica de cifrado que no se puede descifrar si se utiliza correctamente. Ésta técnica posee la caracteristica de Secretismo Perfercto, ésta característica fue definida por Claude Shannon en 1946, éste concepto se define bajo elementos de estadística y probabilidad. Un texto cifrado mantiene un secreto perfecto si el conocimiento del atacante sobre el contenido del mensaje es el mismo antes y después de que el adversario inspeccione el texto cifrado, atacándolo con recursos ilimitados. Es decir, el mensaje no le da al adversario precisamente información sobre el contenido del mensaje. 



### En esta técnica;

Un texto plano se empareja con una clave secreta aleatoria luego, cada bit o carácter del texto plano se cifra combinándolo con el bit o carácter correspondiente del pad utilizando la adición modular y la operación logica XOR. Si la clave es verdaderamente aleatoria, es al menos tan larga como el texto sin formato, nunca se reutiliza en su totalidad o en parte, y se mantiene completamente en secreto, entonces el texto cifrado resultante será imposible de descifrar o romper.[1]
